```python
# 1. Generate a matrix of charts with 3 rows and 4 columns containing the histograms of all the data columns (except ID and quality)
```

```python
import pandas as pd

# Step 1: Read the CSV file into a DataFrame
csv_file_path = '/content/question_3.csv'
df_csv = pd.read_csv(csv_file_path)

# Step 2: Display the contents of the DataFrame
print(df_csv)
```

```python


import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame

# Extract the column names for plotting (excluding 'Id' and 'quality')
columns_to_plot = [col for col in df_csv.columns if col not in ['Id', 'quality']]

# Create a 3x4 grid of subplots for histograms
num_rows = 3
num_cols = 4
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten the 2D array of axes for easy iteration
axes = axes.ravel()

# Generate histograms for each data column and plot in each subplot
for i, column in enumerate(columns_to_plot):
    axes[i].hist(df_csv[column], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

# Hide any empty subplots
for i in range(len(columns_to_plot), num_rows * num_cols):
    axes[i].axis('off')

# Adjust the layout for better visualization
plt.tight_layout()
plt.show()
```
